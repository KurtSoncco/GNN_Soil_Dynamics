# Checking the frequency content of ricker wave

```python
def ricker(freq1, t01, t):
    wave = (1-2.0*(np.pi*freq1*(t-t01))**2)*np.exp(-((np.pi*freq1*(t-t01))**2))
    return wave
```

```python
t = np.arange(0, 15+1e-5, 1e-5)
freq1 = 4.16
t01 = 2.5
ricker_wave = ricker(freq1, t01, t)
```

```python
# FOURIER AMPLITUDE SPECTRUM
from scipy.fft import fft, fftfreq

# Number of sample points
N = len(ricker_wave)
# Sample spacing
T = 1e-5

# Compute the FFT
yf = fft(ricker_wave, n=N, axis=0)
FAS = 2.0/N * np.abs(yf[:N//2])
xf = fftfreq(N, T)[:N//2]


plt.loglog(xf, FAS, label="FAS", linestyle='--', color='black')
plt.axvline(x=freq1, color='red', linestyle='--')
plt.vlines(x=[freq1,3*freq1], ymin=1e-8, ymax=1, colors='green', linestyles=':')
plt.ylim(1e-6, 1)
plt.xlim(1e-1, 1e2)
plt.xlabel("Frequency (Hz)")
plt.ylabel("FAS")
plt.grid(which='both', linestyle=':', linewidth=0.5)
```
